project 2: Number guessing game

```python
secret_number=7
while True:
  guess_number=int(input("enter guess number:"))
  if secret_number==guess_number:
    print("correct guess")
    break
  elif secret_number<guess_number:
    print("to high,try again")
  else:
    print(" to low ,try again")
```
