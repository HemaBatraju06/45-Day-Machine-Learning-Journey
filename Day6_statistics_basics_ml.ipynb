                      Statistics Basics


  Mean - Average of values

```python
marks=[ 10,20,30,40]
mean=sum(marks)/len(marks)
print(mean)
```

Median - middle value

```python
import statistics
data=[10,20,12,5,40,]
print("median=",statistics.median(data))
```

Mode - most frquent or repeated value

```python
import statistics
data=[20,12,20,45,40,]
print("mode=",statistics.mode(data))
```
