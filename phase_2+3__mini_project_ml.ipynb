    Mini_Project  : student Data Analysis System

```python
import numpy as np
# create dataset
import pandas as pd

data = {
    "Name": ["Ram", "Ravi", "Sai", "Kiran", "Ram",
             "Anil", "Priya", "Sita", "Rahul", "Pooja",
             "Vijay", "Keerthi"],
    "Age": [18, 19, 20, 21, 18,
            19, 20, 18, 21, 19,
            20, 18],
    "Marks": [70, 80, 90, 95, 75,
              85, 88, 72, 91, 78,
              83, 87]
}
 #convert to dataframe
df = pd.DataFrame(data)
# save as csv file
df.to_csv("student_data.csv", index=False)
#read csv
df = pd.read_csv("student_data.csv")
# understand datset
print(df.head())
print(df.shape)
print(df.columns)
print(df.info())
# data  cleanining
print(df.isnull().sum())
print(df.duplicated().sum())
# math analysis
print("Average:", df["Marks"].mean())
print("Highest:", df["Marks"].max())
print("Lowest:", df["Marks"].min())
print("Variance:", df["Marks"].var())
print("Standard Deviation:", df["Marks"].std())
#numpy
Marks=np.array(df["Marks"])
print(np.mean(Marks))
```
