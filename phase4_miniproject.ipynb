```python
# ==========================================
# Student Data Analysis & Visualization System
# ==========================================

# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display all columns
pd.set_option("display.max_columns", None)

# ==========================================
# Load Dataset
# ==========================================

df = pd.read_csv("student_data.csv")

# ==========================================
# View Dataset
# ==========================================

print("First 5 Records")
print(df.head())

print("\nLast 5 Records")
print(df.tail())

print("\nShape")
print(df.shape)

print("\nColumns")
print(df.columns)

# ==========================================
# Dataset Information
# ==========================================

print("\nDataset Info")
print(df.info())

print("\nStatistical Summary")
print(df.describe())

# ==========================================
# Missing Values
# ==========================================

print("\nMissing Values")
print(df.isnull().sum())

# Fill Missing Values (Only if present)
if "Marks" in df.columns:
    df["Marks"] = df["Marks"].fillna(df["Marks"].mean())

if "Attendance" in df.columns:
    df["Attendance"] = df["Attendance"].fillna(df["Attendance"].mean())

# ==========================================
# Remove Duplicates
# ==========================================

df = df.drop_duplicates()

print("\nDataset Shape After Removing Duplicates")
print(df.shape)

# ==========================================
# Statistics
# ==========================================

print("\n----- Statistics -----")

print("Mean:", df["Marks"].mean())
print("Median:", df["Marks"].median())
print("Mode:")
print(df["Marks"].mode())

print("Maximum:", df["Marks"].max())
print("Minimum:", df["Marks"].min())

print("Variance:", df["Marks"].var())
print("Standard Deviation:", df["Marks"].std())

# ==========================================
# NumPy Operations
# ==========================================

marks = np.array(df["Marks"])

print("\n----- NumPy -----")

print("Average:", np.mean(marks))
print("Maximum:", np.max(marks))
print("Minimum:", np.min(marks))
print("Standard Deviation:", np.std(marks))

# ==========================================
# Data Visualization using Matplotlib
# ==========================================

# Line Plot
plt.figure(figsize=(8,5))
plt.plot(df["Student_ID"], df["Marks"], marker="o")
plt.title("Student Marks")
plt.xlabel("Student ID")
plt.ylabel("Marks")
plt.grid(True)
plt.show()

# Bar Chart
plt.figure(figsize=(8,5))
plt.bar(df["Department"], df["Marks"])
plt.title("Department vs Marks")
plt.xlabel("Department")
plt.ylabel("Marks")
plt.show()

# Histogram
plt.figure(figsize=(8,5))
plt.hist(df["Marks"], bins=10)
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()

# Scatter Plot
plt.figure(figsize=(8,5))
plt.scatter(df["Attendance"], df["Marks"])
plt.title("Attendance vs Marks")
plt.xlabel("Attendance")
plt.ylabel("Marks")
plt.show()

# Pie Chart
department = df["Department"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    department,
    labels=department.index,
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Department Distribution")
plt.show()

# ==========================================
# Data Visualization using Seaborn
# ==========================================

# Count Plot
plt.figure(figsize=(8,5))
sns.countplot(x="Department", data=df)
plt.title("Students in Each Department")
plt.show()

# Box Plot
plt.figure(figsize=(8,5))
sns.boxplot(x="Department", y="Marks", data=df)
plt.title("Department vs Marks")
plt.show()

# Histogram
plt.figure(figsize=(8,5))
sns.histplot(df["Marks"], bins=10, kde=True)
plt.title("Marks Distribution")
plt.show()

# Scatter Plot
plt.figure(figsize=(8,5))
sns.scatterplot(x="Attendance", y="Marks", data=df)
plt.title("Attendance vs Marks")
plt.show()

# Heatmap (Correlation Matrix)
numeric = df.select_dtypes(include=np.number)

plt.figure(figsize=(8,6))
sns.heatmap(
    numeric.corr(),
    annot=True,
    cmap="coolwarm"
)
plt.title("Correlation Matrix")
plt.show()

# Pair Plot
sns.pairplot(df)
plt.show()

# ==========================================
# Final Insights
# ==========================================

print("\n========== FINAL INSIGHTS ==========")

print("Highest Marks :", df["Marks"].max())
print("Lowest Marks :", df["Marks"].min())
print("Average Marks :", df["Marks"].mean())
print("Average Attendance :", df["Attendance"].mean())

print("\nTop Performing Student")
print(df[df["Marks"] == df["Marks"].max()])

print("\nDepartment Wise Average Marks")
print(df.groupby("Department")["Marks"].mean())

print("\nDepartment Wise Attendance")

print(df.groupby("Department")["Attendance"].mean())

print("\nStudents Scoring Above Average")
print(df[df["Marks"] > df["Marks"].mean()])
pass_percentage = (df["Marks"] >= 40).mean() * 100

print("\nPass Percentage")

print(f"{pass_percentage:.2f}%")
print("\n========== DASHBOARD KPIs ==========\n")

print("Total Students :", len(df))
print("Average Marks :", round(df["Marks"].mean(),2))
print("Highest Marks :", df["Marks"].max())
print("Lowest Marks :", df["Marks"].min())
print("Average Attendance :", round(df["Attendance"].mean(),2))
print("Pass Percentage :", round(pass_percentage,2),"%")

print("\n========== STUDENT DATA ANALYSIS REPORT ==========\n")

print(f"Total Students          : {len(df)}")
print(f"Average Marks           : {df['Marks'].mean():.2f}")
print(f"Highest Marks           : {df['Marks'].max()}")
print(f"Lowest Marks            : {df['Marks'].min()}")
print(f"Average Attendance      : {df['Attendance'].mean():.2f}%")
print(f"Best Department         : {df.groupby('Department')['Marks'].mean().idxmax()}")

print("\nKey Findings:")
print("- Most students scored above the average marks.")
print("- Students with higher attendance generally achieved better marks.")
print("- Department-wise analysis shows differences in academic performance.")
print("- Data visualizations helped identify patterns and trends quickly.")
print("\n========== CONCLUSION ==========\n")

print("""
The Student Data Analysis & Visualization System successfully analyzed
student performance using Python, NumPy, Pandas, Matplotlib, and Seaborn.

The project included data cleaning, statistical analysis, exploratory
data analysis (EDA), visualization, and dashboard storytelling.

The analysis showed that attendance has a positive relationship with
student performance. Dashboard visualizations made it easier to identify
department-wise trends and summarize key insights.

Overall, this project demonstrates a complete data analysis workflow,
from loading a CSV dataset to presenting meaningful insights through
charts, reports, and conclusions.
""")
```
