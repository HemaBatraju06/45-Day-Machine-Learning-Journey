        Data cleaning ( missing values and Duplicate values)

 Data cleaning : It is a process of fixing incorrect , missing or duplicate data.


 Missing values: values that are not available in data


```python
# create dataset
import pandas as pd
data={
    "Name": ["x",'Y','Z'],
    "Age": [6,None,9]

}
df=pd.DataFrame(data)
print(df)
```

```python
#check missing values
print(df.isnull())
```

```python
#count missing values
print(df.isnull().sum())
```

```python
# remove missing value
print(df.dropna())
```

Duplicate values :  repeated record in a dataset

```python
#create dataset

import pandas as pd
data={
    "Name":["XX","YY","XX"],
    "Age":[6,7,6]
}
df=pd.DataFrame(data)
print(df)

```

```python
#check duplicate values
print(df.duplicated())
```

```python
# remove duplicate values
print(df.drop_duplicates())
```

```python
import pandas as pd
data={
    "Name":["A","B","B","C"],
    " Age" : [18,None,None,20],
    "Marks": [70,80,80,90]
}
df=pd.DataFrame(data)
print(df)
print(df.isnull())
print(df.isnull().sum())
print(df.dropna())
print(df.duplicated())
print(df.drop_duplicates())
```
