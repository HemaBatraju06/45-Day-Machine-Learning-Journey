                    DASHBOARD STORYTELLING

Dashboard - collecttion of charts and KPIs on one screen.

Storytelling : eplain the what charts mean .

KPIs: Important numbers (Average,Highest,Lowest,Total).

Insights: Useful conclusions from the data.

Conclusion: Final summary of the analysis.

MINI PROJECT :  

```python
# =====================================
# Dashboard Storytelling Example
# =====================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create Sample Dataset
data = {
    "Student":["Ravi","Priya","Rahul","Sneha","Akhil"],
    "Department":["AIML","CSE","AIML","ECE","CSE"],
    "Marks":[90,85,70,95,60],
    "Attendance":[95,92,80,98,75]
}

df = pd.DataFrame(data)

# =====================================
# KPI's
# =====================================

print("========== DASHBOARD ==========\n")

print("Total Students :", len(df))
print("Average Marks :", round(df["Marks"].mean(),2))
print("Highest Marks :", df["Marks"].max())
print("Lowest Marks :", df["Marks"].min())
print("Average Attendance :", round(df["Attendance"].mean(),2))

# =====================================
# Charts
# =====================================

# Department-wise Average Marks
plt.figure(figsize=(6,4))
df.groupby("Department")["Marks"].mean().plot(kind="bar", color=["skyblue","orange","green"])
plt.title("Department-wise Average Marks")
plt.xlabel("Department")
plt.ylabel("Average Marks")
plt.show()

# Attendance vs Marks
plt.figure(figsize=(6,4))
sns.scatterplot(x="Attendance",y="Marks",data=df)
plt.title("Attendance vs Marks")
plt.show()

# Marks Distribution
plt.figure(figsize=(6,4))
sns.histplot(df["Marks"],bins=5,kde=True)
plt.title("Marks Distribution")
plt.show()

# Correlation Heatmap
plt.figure(figsize=(4,3))
sns.heatmap(df[["Marks","Attendance"]].corr(),annot=True,cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()

# =====================================
# Insights
# =====================================

print("\n========== INSIGHTS ==========\n")

print("1. Average Marks :", round(df["Marks"].mean(),2))

best_department = df.groupby("Department")["Marks"].mean().idxmax()
print("2.", best_department, "department has the highest average marks.")

top_student = df.loc[df["Marks"].idxmax(),"Student"]
print("3. Top Performing Student :", top_student)

print("4. Students with higher attendance generally score better.")

# =====================================
# Report
# =====================================

print("\n========== REPORT ==========\n")

print("""
Student Performance Report

• Total Students : 5

• Average Marks : 80

• Highest Marks : 95

• Lowest Marks : 60

• Best Department : ECE

• Higher attendance is positively related to better marks.

Overall student performance is good.
""")

# =====================================
# Conclusion
# =====================================

print("========== CONCLUSION ==========\n")

print("""
The dashboard summarizes student performance using KPIs and visualizations.
The analysis shows that students with higher attendance tend to score higher marks.
Dashboard storytelling helps convert raw data into meaningful insights for better decision-making.
""")
```
