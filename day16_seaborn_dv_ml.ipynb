Seaborn :

     A python library is used to creating beautiful satistical and attractive visualizations.


```python
# importing of seaborn
import seaborn as sns
import matplotlib
```

Heatmap :
        
        A graph that represents values using color, commonly used to visualize a correlation matrix.

```python
import seaborn as sns
import matplotlib.pyplot as plt
data=[
    [1,2,3],
    [6,5,9],
    [9,6,3]
]
sns.heatmap(data,annot=True)
plt.show
```

Distribution plot :
     
     
      shows how data is distributed

```python
import seaborn as sns
import matplotlib.pyplot as plt

marks = [65,70,72,75,76,78,80,95]

sns.histplot(marks, kde=True)

plt.show()
```

Box plot :

      used to summarize the data and detect errors

```python
import seaborn as sns
import matplotlib.pyplot as plt

marks = [70,72,74,75,76,78,80,98]

sns.boxplot(x=marks)

plt.show()
```

pair plot:

    that compares relationships between all numerical features

```python
import pandas as pd
import seaborn as sns

data = {
    "Age":[18,19,20,21,22],
    "Marks":[70,75,80,90,95],
    "Attendance":[80,82,85,90,95]
}

df = pd.DataFrame(data)

sns.pairplot(df)
```

Correlation matrix: Table showing stength and direction of relatioonship between numerical values.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "Age":[18,19,20,21,22],
    "Marks":[70,75,80,90,95],
    "Attendance":[80,82,85,90,95]
}

df = pd.DataFrame(data)

correlation = df.corr(numeric_only=True)

sns.heatmap(correlation, annot=True)

plt.show()
```

Count plot :


   
     that displays how times each category appears.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "Gender":["Male","Female","Male","Male","Female","Female","Male"]
}

df = pd.DataFrame(data)

sns.countplot(x="Gender", data=df)

plt.show()
```
