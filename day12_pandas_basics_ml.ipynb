       Pandas_Basics : (series,dataframe)

---



pandas : Tables with rows and columns of data


Series: single columns of data


```python
#creatind series
import pandas as pd
marks=pd.Series([30,35,40])
print(marks)
```

```python
#accessing series elements
print(marks[0])
```

Dataframe : complete table with rows and columns

```python
#creating dataframe
import pandas as pd
data={
  "Name":["X","Y","Z"],
    "Marks":[30,40,50]
}
df=pd.DataFrame(data)
print(df)

```

```python
# view column
print(df["Name"])
```

```python
# multiple_columns
print(df[["Name","Marks"]])
```

```python
# shape in dataframe
print(df.shape)
```

```python
# columns in dataframe
print(df.columns)
```

```python
#Head in dataframe
print(df.head())
```
