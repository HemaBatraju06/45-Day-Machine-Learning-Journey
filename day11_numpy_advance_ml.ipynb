```python
import numpy as np
a=np.array([2,5,7,9,0,1])
print(a)
```

```python
#addition
print(a+5)
```

```python
#substraction
print(a-7)
```

```python
#multiplication
print( a*3)
```

```python
#Division
print(a/2)
```

```python
# array to array operations
import numpy as np
a=np.array([2,5,9])
b=np.array([1,2,3])
print("add:",a+b)
print("sub:",a-b)
print("mul:",a*b)
print("div:",a/b)
```

```python
#mean function
a=np.array([2,5,9])
print(np.mean(a))
```

```python
#sum function
a=np.array([2,5,9])
print(np.sum(a))
```

```python
#max func
a=np.array([2,5,9])
print(np.max(a))
```

```python
#min func
a=np.array([2,5,9])
print(np.min(a))
```

Mini task : take student marks and find total ,avg ,highest , lowest marks

```python
import numpy as np
marks=np.array([60,70,80,90,95])
print("total marks:",np.sum(marks))
print("avg marks:",np.mean(marks))
print("highest marks:",np.max(marks))
print("lowest marks:",np.min(marks))
```
