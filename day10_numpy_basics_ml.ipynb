                              **Numpy Basics**

Array : Collection of values stored in a single variable

```python
import numpy as np
data=np.array([1,2,3,4,5]) # array created
print(data)
```

Indexing : access the elements

```python
import numpy as np
data=np.array([1,2,3,4,5]) # array created
print(data[2])
```

```python
# negative indexing

import numpy as np
data=np.array([1,2,3,4,5]) # array created
print(data[-3])
```

Silicing : Access multiple elements


```python
import numpy as np
data=np.array([1,2,3,4,5]) # array created
print(data[:4])
```
