                  CSV  Dataset Handling

CSV file : It stores data in rows and columns.


    csv- comma seperated values  






```python
import pandas as pd

data = {
    "Name": ["Ram", "Ravi", "Sai"],
    "Marks": [70, 80, 90]
}

df = pd.DataFrame(data)

df.to_csv("student_data.csv", index=False)

```

Reading CSV file : this loads the datasets into a dataframe.
