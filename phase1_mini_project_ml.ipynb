Mini_project - " Smart Household Expenses Tracker"

```python
expenses = []
def add_expense():
    category = input("Enter Category: ")
    amount = float(input("Enter Amount: "))
    expense = {
        "category": category,
        "amount": amount
    }
    expenses.append(expense)
    print("Expense Added Successfully")
def view_expenses():
    if len(expenses) == 0:
        print("No Expenses Found")
    else:
        print("\n--- Expense Records ---")
        for expense in expenses:
            print(
                expense["category"],
                "- ₹",
                expense["amount"]
            )
def total_spending():
    total = 0
    for expense in expenses:
        total += expense["amount"]
    print("Total Spending = ₹", total)
def category_wise_expenses():
    if len(expenses) == 0:
        print("No Expenses Found")
    else:
        print("\n--- Category Wise Expenses ---")
        categories = {}
        for expense in expenses:
            category = expense["category"]
            amount = expense["amount"]
            if category in categories:
                categories[category] += amount
            else:
                categories[category] = amount
        for category, amount in categories.items():
            print(category, "- ₹", amount)
while True:
    print("\n=====  Smart Household Expense Tracker =====")
    print("1. Add Daily Expense")
    print("2. View Expense Records")
    print("3. Calculate Total Spending")
    print("4. Category Wise Expenses")
    print("5. Exit")
    choice = input("Enter Choice: ")
    if choice == "1":
        add_expense()
    elif choice == "2":
        view_expenses()
    elif choice == "3":
        total_spending()
    elif choice == "4":
        category_wise_expenses()
    elif choice == "5":
        print("Thank You!")
        break
    else:
        print("Invalid Choice")
```
