                Linear Algebra(vectors & Martrices)


Vector - is a collection of numbers in single row and column

```python
marks=[15,20,30]
print( marks)
```

Access vector elements:


```python
vector=[20,40,60]
print(vector[1])
print(vector[2])
```

Martrix -  is a collection of rows and columns (vectors)

```python
matrix=[
    [ 10,30,50],
    [20,40,60]
]
print(matrix)
```

Access matrix elements-

```python
matrix=[
    [ 10,30,50],
    [20,40,60]
]
print(matrix[0][0])
print(matrix[1][1])
print(matrix[1][2])
print(matrix[1][0])
```

Matrix Operations :

    Addition
    Subtraction
    Multiplication (Most Important in ML)
    Scalar Multiplication

```python
#additon

A=[[2,4],[3,6]]
B=[[3,5],[4,7]]
result = [
    [A[0][0] + B[0][0], A[0][1] + B[0][1]],
    [A[1][0] + B[1][0], A[1][1] + B[1][1]]
]
print(result)
```
