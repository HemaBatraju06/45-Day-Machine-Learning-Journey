                  Matplotlib(Basic charts)



Matplotlib : It is a pyhon library used to create graphs and chats for data visualization.


```
import matplotlib.pyplot as plt
```
pyplot - is used to creste graphs






Line chart : It shows trends over time


````python
import matplotlib.pyplot as plt

students = ["Ram", "Ravi", "Sai", "Anil", "Priya"]
marks = [70, 85, 90, 80, 95]

plt.plot(students, marks)

plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
````

Bar chart: compare values using bars

````python
import matplotlib.pyplot as plt

students = ["Ram", "Ravi", "Sai", "Anil", "Priya"]
marks = [70, 85, 90, 80, 95]

plt.bar(students, marks)

plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
````

Pie chart:

````python
import matplotlib.pyplot as plt

students = ["Ram", "Ravi", "Sai", "Anil", "Priya"]
marks = [70, 85, 90, 80, 95]

plt.pie(marks, labels=students, autopct="%1.1f%%")

plt.title("Student Marks Distribution")

plt.show()
````

````python
# line chart for monthly sales
import matplotlib.pyplot as plt
months=["jan","feb","mar","apr","may","jun","jul","aug","sep","oct","nov","dec"]
sales=[1000,2000,5000,2500,1500,4500,10000,7000,8000,7500,3500,3000]
plt.plot(months,sales)
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
````

````python

#bar chart for comparing subject marks
import matplotlib.pyplot as plt
subject=["subj1","subj2","subj3","subj4","subj5"]
marks=[99,80,95,60,75]
plt.bar(subject,marks)
plt.title("Subject Marks")
plt.xlabel("Subject")
plt.ylabel("Marks")
plt.show()
````

MiniProject:

````python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    "Name": ["Ram", "Ravi", "Sai", "Kiran", "Ram",
             "Anil", "Priya", "Sita", "Rahul", "Pooja",
             "Vijay", "Keerthi"],
    "Age": [18, 19, 20, 21, 18,
            19, 20, 18, 21, 19,
            20, 18],
    "Marks": [70, 80, 90, 95, 75,
              85, 88, 72, 91, 78,
              83, 87]
}
 #convert to dataframe
df = pd.DataFrame(data)
# save as csv file
df.to_csv("student_data.csv", index=False)
#read csv
df = pd.read_csv("student_data.csv")
#line chart
plt.plot(df["Name"], df["Marks"])
plt.title(" Line chart-Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
#bar chart
plt.bar(df["Name"], df["Marks"])
plt.title("Bar chart-Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
#pie chart
plt.pie(df["Marks"], labels=df["Name"], autopct="%1.1f%%")
plt.title("Pie chart-Student Marks")
plt.show()
````
